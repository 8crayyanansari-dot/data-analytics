```python
# 1- a) Importing important libraries for data cleaning 
#    b) Reading the csv file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dt = pd.read_csv('dirty_healthy_diet_calorie_intake.csv')
```

```python
# 2- Checking whether the file is properly readable or not
print("<---The first 5 data--->")
print(dt.head(5).to_string())
print("<---The last 5 data--->")
print(dt.tail(5).to_string())
```

```python
# 3- Coutning of null attributes in each column
total_nulls = dt.isnull().sum()
print("Total null attributes in each column--->")
print(total_nulls)
```

```python
# 4- Filling null values of 'Age' column 
age_median = dt['Age'].median()
dt['Age'] = dt['Age'].fillna(age_median)
print(dt['Age'])
```

```python
# 5- Filling null values of 'Gender' column 
gender_mode = dt['Gender'].mode()[0]
dt['Gender'] = dt['Gender'].fillna(gender_mode)
print(dt['Gender'])
```

```python
# 6- Filling null values of 'Weight_kg' column 
Weight_mean = dt['Weight_kg'].mean()
dt['Weight_kg'] = dt['Weight_kg'].fillna(Weight_mean)
print(dt['Weight_kg'])
```

```python
# 7- Filling null values of 'Activity_level' column 
level_mode = dt['Activity_Level'].mode()[0]
dt['Activity_Level'] = dt['Activity_Level'].fillna(level_mode)
print(dt['Activity_Level'])
```

```python
# 8- Filling null values of 'Daily_Calorie_Consumed' column 
calorie_mean = dt['Daily_Calorie_Consumed'].mean()
dt['Daily_Calorie_Consumed'] = dt['Daily_Calorie_Consumed'].fillna(calorie_mean)
print(dt['Daily_Calorie_Consumed'])

```

```python
# 9- Filling null values of 'Diet_Type' column 
diet_mode = dt['Diet_Type'].mode()[0]
dt['Diet_Type'] = dt['Diet_Type'].fillna(diet_mode)
print(dt['Diet_Type'])
```

```python
# 10- Saving changes as new named file
dt.to_csv('C:/Users/DELL/OneDrive/Desktop/ASSIGNMENTS/Project 2/Cleaned_data2.csv',index=False)
```

```python
# 11- Creating a graph for visualization
top_10_bmi = dt.sort_values(by='BMI', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_10_bmi, x='Weight_kg', y='BMI', s=150, color='red')
sns.regplot(data=top_10_bmi, x='Weight_kg', y='BMI', scatter=False, color='blue', label='Trend Line')
plt.title('Relationship: Weight vs BMI (Top 10 Highest BMI)', fontsize=14, fontweight='bold')
plt.xlabel('Weight (kg)')
plt.ylabel('BMI')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
