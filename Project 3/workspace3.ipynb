```python
# 1- a) Importing important libraries for data cleaning 
#    b) Reading the csv file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dt = pd.read_csv('messy_Student_Social_Media_Mental_Health.csv')
```

```python
# 2- Checking whether the file is properly readable or not
print("<---The first 5 data--->")
print(dt.head(5).to_string())
print("<---The last 5 data--->")
print(dt.tail(5).to_string())
```

```python
# 3- Coutning of null attributes in each column
total_nulls = dt.isnull().sum()
print("Total null attributes in each column--->")
print(total_nulls)
```

```python
# 4- Dropping all null attributes
dt.dropna(inplace=True)
print(dt.to_string())
```

```python
# 5- Solving negative values 
for i in dt.index:
    if dt.at[i,'Physical_Activity_Hours'] < 0:
        dt.at[i,'Physical_Activity_Hours'] = 0
    else:
        pass
    
```

```python
# 6- Fixing data types
dt['Age'] = dt['Age'].astype('int64')
dt['Daily_Unlocks'] = dt['Daily_Unlocks'].astype('int64') 
print(dt.head(5).to_string())
```

```python
# 7- Reloading the 'Stress_Level' column data as it became null
raw_data = pd.read_csv("messy_Student_Social_Media_Mental_Health.csv") # Replace with your exact filename if different
dt['Stress_Level'] = raw_data['Stress_Level']
```

```python
# 8- Saving changes into another named file
dt = dt.reset_index(drop=True)
dt.to_csv(r'C:\Users\DELL\OneDrive\Desktop\ASSIGNMENTS\Project 3\Cleaned_data3.csv',index=True)
```

```python
# 9- Graph for data visualization
# 9- Simple graph for data visualization

plt.figure(figsize=(6,4))

plt.hist(dt["Mental_Health_Score"])

plt.title("Mental Health Score Distribution")
plt.xlabel("Mental Health Score")
plt.ylabel("Frequency")

plt.show()
```
